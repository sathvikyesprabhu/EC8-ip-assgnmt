Q.1 write python code  from scratch to convert color image (Rose.jpg and Lenna.tif)  to grayscale image using  a def function and list of comprehension method and plot processing time comparison with  OpenCV and forloop methods

```python
# To Grayscale
import cv2
import matplotlib.pyplot as plt
import time
import numpy as np

# Formula used intensity = 0.21*r + 0.71*g + 0.07*b

# Reading the image
img=cv2.imread('Rose.jpg')
img1=img2=img3=img4=img
plt.imshow(img)
plt.title('Input BGR image')
plt.show()
s=img.shape
print(s)


# Method 1 Def function
def colortogray(img):
    s=img.shape
    for i in range(s[0]):
        for j in range(s[1]):
            img1[i][j]= 0.21*img[i][j][2] + 0.71*img[i][j][1] + 0.07*img[i][j][0]
    return img1
t10=time.time()
img1=colortogray(img)
t11=time.time()
print("Time taken: ",t11-t10, "s")

plt.imshow(img1, cmap="gray")
plt.title('Method 1: Grayscale image')
plt.show()

# Method 2 List of comprehension

t20=time.time()
img2= [[0.21*img[i][j][2] + 0.71*img[i][j][1] + 0.07*img[i][j][0] for j in range(s[1])] for i in range(s[0])]
t21=time.time()
print("Time taken: ",t21-t20, "s")
        
plt.imshow(img2,cmap="gray")
plt.title('Method 2: Grayscale image')
plt.show()

# Method 3 OpenCV
t30=time.time()
gray_img=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
t31=time.time()
print("Time taken: ",t31-t30, "s")
        
plt.imshow(gray_img, cmap="gray")
plt.title('Method 3: Grayscale image')
plt.show()

# Method 4 For loop
t40=time.time()
for i in range(s[0]):
    for j in range(s[1]):
        img4[i][j]= 0.21*img[i][j][2] + 0.71*img[i][j][1] + 0.07*img[i][j][0]
t41=time.time()
print("Time taken: ",t41-t40, "s")
        
plt.imshow(img4, cmap="gray")
plt.title('Method 4: Grayscale image')
plt.show()

# Write 
# cv2.imwrite('grayscale.jpg',gray_img)

# Comparison
x=["Method 1","Method 2","Method 3","Method 4"]
t=[t11-t10,t21-t20,t31-t30,t41-t40]
plt.bar(x,t,color="blue")
plt.xlabel("Method")
plt.ylabel("Processing Time")
plt.title("Comparison of methods")
plt.show()
```

Q 2.Write python code  from scratch to convert Grayscale  image to Binary images using a def function and list of comprehension method and  plot processing time comparison with OpenCV and forloop methods

```python
# To binary
import cv2
import matplotlib.pyplot as plt
import time
import numpy as np


# Reading the image
img_color=cv2.imread('Rose.jpg')
img=cv2.cvtColor(img_color,cv2.COLOR_BGR2GRAY)
plt.imshow(img,cmap='gray')
plt.title('Input Grayscale image')
plt.show()
s=img.shape
print(s)
img1=img2=img3=img4=np.zeros(s)
print(img.dtype)


# Method 1 Def function
def graytobinary(img):
    s=img.shape
    for i in range(s[0]):
        for j in range(s[1]):
            if img[i][j]<=127:
                img1[i][j]=0
            else:
                img1[i][j]=1
    return img1

t10=time.time()
img1=graytobinary(img)
t11=time.time()
print("Time taken: ",t11-t10, "s")

plt.imshow(img1,cmap='gray')
plt.title('Method 1: Binary image')
plt.show()

# Method 2 List of comprehension

t20=time.time()
img2= [[0 if img[i][j]<=127 else 1 for j in range(s[1])] for i in range(s[0])] 
t21=time.time()
print("Time taken: ",t21-t20, "s")
        
plt.imshow(img2,cmap="gray")
plt.title('Method 2: Binary image')
plt.show()

# Method 3 OpenCV
t30=time.time()
(thresh,binary_img)=cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)
t31=time.time()
print("Time taken: ",t31-t30, "s")
        
plt.imshow(binary_img,cmap='gray')
plt.title('Method 3: Binary image')
plt.show()

# Method 4 For loop
t40=time.time()

for i in range(s[0]):
    for j in range(s[1]):
        if img[i][j]<=127:
            img4[i][j]=0
        else:
            img4[i][j]=1
            
t41=time.time()
print("Time taken: ",t41-t40, "s")
        
plt.imshow(img4,cmap='gray')
plt.title('Method 4: Binary image')
plt.show()

# Write 
# cv2.imwrite('grayscale.jpg',gray_img)

# Comparison
x=["Method 1","Method 2","Method 3","Method 4"]
t=[t11-t10,t21-t20,t31-t30,t41-t40]
plt.bar(x,t,color="blue")
plt.xlabel("Method")
plt.ylabel("Processing Time")
plt.title("Comparison of methods")
plt.show()
```

Q3.  Write python code from scratch to  add two images (image1.jpg and image2.jpg) using def function and list of comprehension method and plot processing time comparison with OpenCV and forloop methods.

```python
# Add two images
import cv2
import matplotlib.pyplot as plt
import time
import numpy as np

# Read
img1=cv2.imread('input1.jpg',0)
plt.imshow(img1,"gray")
plt.title('Input 1')
plt.show()
s1=img1.shape
print("Image 1 shape: ",s1)

img2=cv2.imread('input2.jpg',0)
plt.imshow(img2,"gray")
plt.title('Input 2')
plt.show()
s2=img2.shape
print("Image 2 shape: ",s2)
s=s1

# Method 1 Def function
img=np.zeros(s)
def addimages(img1,img2):
    s=img1.shape
    for i in range(s[0]):
        for j in range(s[1]):
            img[i][j]= min(int(img1[i][j]) + int(img2[i][j]) , 255)
    return img

t10=time.time()
img=addimages(img1,img2)
t11=time.time()
print("Time taken: ",t11-t10, "s")

plt.imshow(img, cmap="gray")
plt.title('Method 1')
plt.show()

# Method 2 List of comprehension
img=np.zeros(s)
t20=time.time()
img= [[min(int(img1[i][j]) + int(img2[i][j]),255) for j in range(s[1])] for i in range(s[0])]
t21=time.time()
print("Time taken: ",t21-t20, "s")
        
plt.imshow(img,cmap="gray")
plt.title('Method 2')
plt.show()

# Method 3 OpenCV
img=np.zeros(s)
t30=time.time()
# img=cv2.addWeighted(img1,0.5,img2,0.5,0)
img=cv2.add(img1,img2)
t31=time.time()
print("Time taken: ",t31-t30, "s")
        
plt.imshow(img, cmap="gray")
plt.title('Method 3')
plt.show()

# Method 4 For loop
img=np.zeros(s)
t40=time.time()
for i in range(s[0]):
    for j in range(s[1]):
        img[i][j]= min(int(img1[i][j]) + int(img2[i][j]),255)
t41=time.time()
print("Time taken: ",t41-t40, "s")
        
plt.imshow(img, cmap="gray")
plt.title('Method 4')
plt.show()

# # Write 
# # cv2.imwrite('grayscale.jpg',gray_img)

# Comparison
x=["Method 1","Method 2","Method 3","Method 4"]
t=[t11-t10,t21-t20,t31-t30,t41-t40]
plt.bar(x,t,color="blue")
plt.xlabel("Method")
plt.ylabel("Processing Time")
plt.title("Comparison of methods")
plt.show()
```


Q4.  Write python code from scratch to subtract two images(image1.jpg and image2.jpg) using def function and list of comprehension method and plot processing time comparison with OpenCV and forloop methods.

```python
# Subtract two images
import cv2
import matplotlib.pyplot as plt
import time
import numpy as np

# Read
img1=cv2.imread('input1.jpg',0)
plt.imshow(img1,"gray")
plt.title('Input 1')
plt.show()
s1=img1.shape
print("Image 1 shape: ",s1)

img2=cv2.imread('input2.jpg',0)
plt.imshow(img2,"gray")
plt.title('Input 2')
plt.show()
s2=img2.shape
print("Image 2 shape: ",s2)
s=s1

# Formula used: result= img1 - img2

# Method 1 Def function
def subtractimages(img1,img2):
    s=img1.shape
    for i in range(s[0]):
        for j in range(s[1]):
            img[i][j]= int(img1[i][j])-int(img2[i][j])
    return img

t10=time.time()
img=subtractimages(img1,img2)
t11=time.time()
print("Time taken: ",t11-t10, "s")

plt.imshow(img, cmap="gray")
plt.title('Method 1')
plt.show()

# Method 2 List of comprehension
t20=time.time()
img= [[int(img1[i][j]) - int(img2[i][j]) for j in range(s[1])] for i in range(s[0])]
t21=time.time()
print("Time taken: ",t21-t20, "s")
        
plt.imshow(img,cmap="gray")
plt.title('Method 2')
plt.show()

# Method 3 OpenCV
t30=time.time()
img=cv2.subtract(img1,img2)
t31=time.time()
print("Time taken: ",t31-t30, "s")
        
plt.imshow(img, cmap="gray")
plt.title('Method 3')
plt.show()

# Method 4 For loop
img=np.zeros(s)
t40=time.time()
for i in range(s[0]):
    for j in range(s[1]):
        img[i][j]= int(img1[i][j]) - int(img2[i][j])
t41=time.time()
print("Time taken: ",t41-t40, "s")
        
plt.imshow(img, cmap="gray")
plt.title('Method 4')
plt.show()

# # Write 
# # cv2.imwrite('grayscale.jpg',gray_img)

# Comparison
x=["Method 1","Method 2","Method 3","Method 4"]
t=[t11-t10,t21-t20,t31-t30,t41-t40]
plt.bar(x,t,color="blue")
plt.xlabel("Method")
plt.ylabel("Processing Time")
plt.title("Comparison of methods")
plt.show()
```
